```python
import numpy as np
#from sklearn.naive_bayes import MultinomialNB

import pandas as pd

df = pd.read_csv('dataset.csv')
df.head()
```

```python
df.shape
```

```python
import string
df['Label'].value_counts()
```

```python
from nltk.corpus import stopwords

stopWords = set(stopwords.words('english'))
#print(stopWords)

import string
def text_cleaning(a):
 remove_punctuation = [char for char in a if char not in string.punctuation]
 #print(remove_punctuation)
 remove_punctuation=''.join(remove_punctuation)
 #print(remove_punctuation)   
 return [word for word in remove_punctuation.split() if word.lower() not in stopwords.words('english')]
```

```python

from sklearn.feature_extraction.text import CountVectorizer
bow_transformer = CountVectorizer(analyzer=text_cleaning).fit(df['Error_List']) 
  
#print(len(bow_transformer.vocabulary_))   
bow_transformer.vocabulary_
```

```python
tokens = bow_transformer.get_feature_names()
print(tokens)

```

```python
title_bow = bow_transformer.transform(df['Error_List'])

print(title_bow)
```

```python
X = title_bow.toarray()
print(X)

 
X.shape  # 30,46  # 46 seperate words in our dataset and 30 rows
```

```python
# If you want you can store it in a dataframe also
dd=pd.DataFrame(data=title_bow.toarray(),columns=tokens)
dd
```

```python
from sklearn.model_selection import train_test_split
#Splitting dataset into 90% training set and 10% test set
x_train, x_test, y_train, y_test = train_test_split(X, df.Label,test_size=0.1, random_state=50)
```

```python
print(x_train.shape) # 27 rows and 46 columns as we have 46 different vocalublary 
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB().fit(x_train,y_train)
```

```python
prediction = model.predict(x_test)
print(prediction)
```

```python
from sklearn.metrics import accuracy_score
#Printing accuracy of the our model
accuracy_score(prediction,y_test)
```
